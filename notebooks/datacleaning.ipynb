```
df.columns = df.columns.str.strip().str.lower().str.replace(" ","_")
```

```
print(df.columns.tolist())
```

```
print(df['customer_id'])
```

```
# See which columns have missing values
print(df.isnull().sum())

# Option A — drop rows where critical columns are missing
df = df.dropna()

# Option B — fill missing values for less critical columns
df['quantity'] = df["quantity"].fillna("Unknown")
```

```
# Convert order_date and ship_date to proper date format
df['date'] = pd.to_datetime(df["date"])

# Make sure sales and profit are numbers
df['price_per_unit'] = pd.to_numeric(df['price_per_unit'], errors="coerce")
df['quantity'] = pd.to_numeric(df['quantity'], errors="coerce")
df['total_amount'] = pd.to_numeric(df['total_amount'], errors="coerce")
```

```
# Recalculate what the total SHOULD be
df['calculated_total'] = df['quantity'] * df['price_per_unit']

# Find rows where the stored value doesn't match
mismatches = df[df['calculated_total'] != df['total_amount']]

print(f'Rows where total_amount is wrong: {len(mismatches)}')
```

```
# Overwrite wrong values with the correct calculation
df['total_amount'] = df['calculated_total']
print('Fixed!')
```

```
df.drop(columns=['calculated_total'], inplace=True)
```

```
df['month'] = df['date'].dt.month          # 1, 2, 3 ... 12
df['month_name'] = df['date'].dt.strftime('%B')  # January, February...
df['year'] = df['date'].dt.year            # 2023
```

```
def age_group(age):
    if age < 25:
        return 'Under 25'
    elif age < 35:
        return '25–34'
    elif age < 45:
        return '35–44'
    elif age < 55:
        return '45–54'
    else:
        return '55+'

df['age_group'] = df['age'].apply(age_group)
```

```
print("Before:", df.shape)
df = df.drop_duplicates()
print("After:", df.shape)  # should be same or fewer rows
```
